```python
import pandas as pd
from datetime import datetime 
from pandas import DataFrame as df
from matplotlib import style,pyplot as plot


#data fetching
data=pd.read_csv('covid_19_india.csv',index_col=1,parse_dates=True,dayfirst=True)
data.head()
df=pd.DataFrame(data)

start_date = '2020-06-01'
end_date   = '2020-06-30'
df=df.loc[start_date : end_date]

df['State&UTs']
state=df.loc[(df['State&UTs']=='Maharashtra')]

df=pd.DataFrame(state)

df=df.reset_index()
df.get(['Sno', 'Date', 'State&UTs', 'Cured', 'Deaths', 'Confirmed'])

for data in df.index: 
    df['Date'][data], df['State&UTs'][data], df['Cured'][data], df['Deaths'][data], df['Confirmed'][data]
df
```

```python
#plotting line graph
dateList = [i.strftime("%d") for i in df['Date'].tolist()]


plot.plot(dateList,df['Confirmed'].tolist(),'g', label='Confirmed Cases', linewidth =2)
plot.plot(dateList,df['Cured'].tolist(),'m', label='Recovered Cases', linewidth =2)
plot.plot(dateList,df['Deaths'].tolist(),'r', label='Deceased Cases', linewidth =2)
          

plot.title('COVID-19 DATA IN MAHARASHTRA ')
plot.xlabel('Number Of Days in June')
plot.ylabel('Number Of People')
plot.legend()

plot.grid(True)
plot.legend()
plot.savefig('pic.jpg')
plot.show()
```
