```python
# required libraries
import pandas as pd
import numpy as np
from datetime import datetime 
from pandas import DataFrame as df
from matplotlib import style,pyplot as plot
import seaborn as sns
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
```

```python
#data fetching
data1=pd.read_csv('state_wise.csv')
#data1.head()

#df1=pd.DataFrame(data1)

df1=data1.drop([0],axis=0) #dropping irrelevant row
df1
#selecting relevant columns
table = df1.pivot_table(index=['State'],values=['Confirmed','Recovered','Deaths','Active'])
table
```

```python
infected = sum(table["Confirmed"])
recovered = sum(table["Recovered"])
deceased = sum(table["Deaths"])

labels = ['Infected', 'Recovered', 'Deceased']
sizes = [infected, recovered, deceased]
colors = ['cornflowerblue', 'yellowgreen', 'red']
explode = (0, 0, 0) 
plot.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plot.title("Percentage distribution Of Covid-19 Infected, Recovered, Deceased in India till now\n")
plot.axis('equal')
plot.show()
```

```python
index1=["AN","AP","Arn.P","AS","BH","CHD","CT","D&N and D&D","DL","GA","GJt","HR","HP","J&K","JH","KR","KL","LD","LA","MP","MH","MN","ML","MZ","NL","OR","PY","PB","RJ","SK","StateUnas","TN","TG","TR","UP","UK","WB"]
x=np.arange(len(index1))
# set width of bar
barWidth = 0.40
plot.figure(figsize=(30,20),dpi=100) 
# set height of bar
bars1 = table["Confirmed"]
bars2 = table["Recovered"]
bars3 = table["Deaths"]
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
# Make the plot
plot.bar(r1, bars1, color='#7f6d5f',width=barWidth,edgecolor='white', label='Confirmed')
plot.bar(r2, bars2, color='#557f2d',width=barWidth,edgecolor='white', label='Recovered')
plot.bar(r3, bars3, color='#2d7f5e',width=barWidth,edgecolor='white', label='Deaths')
 
# Add xticks on the middle of the group bars
plot.xlabel('States', fontweight='bold')
plot.ylabel('Number of Personals', fontweight='bold')
plot.title('Bar graph showing Infected, Recovered, Deaths of COVID-19 in India\n', fontweight='bold')


plot.xticks([r + barWidth for r in range(len(bars1))],index1)
 
# Create legend & Show graphic
plot.legend()
plot.show()

#tried horizontal bar error occuring StrCategoryConverter- unable to convert states in y-axis

```

```python
#data fetching
data2=pd.read_csv('covid_19_india.csv',parse_dates=['Date'],index_col= ['Date'],na_values=['999.99'])
precip_jan_june_2020 = data2['2020-01-01':'2020-06-30']

precip_jan_june_2020.head()




#data.head()
df=pd.DataFrame(precip_jan_june_2020)

df.drop('Sno',axis=1,inplace=True)

df['State&UTs']
state=pd.DataFrame(df.loc[(df['State&UTs']=='Jharkhand')])

state
```

```python

fig, ax = plot.subplots(figsize=(15, 15))

# Add x-axis and y-axis
ax.bar(data2.index.values,data2['Confirmed'],color='blue')
ax.bar(data2.index.values,data2['Deaths'],color='red')
ax.bar(data2.index.values,data2['Cured'],color='green')

# Set title and labels for axes
ax.set(xlabel="Month",ylabel="Number of people",title="Time series graph- Month-wise Total Covid-19 cases in Jharkhand\n")

#define the date format
date_form = DateFormatter("%m-%d")
ax.xaxis.set_major_formatter(date_form)


plot.show()
```

```python
table = state.pivot_table(index='Date',values=['Cured','Deaths','Confirmed'])
table
```

```python
#plotting line graph
dateList = [i.strftime("%d") for i in table['Date'].tolist()]


plot.plot(dateList,table['Confirmed'].tolist(),'g', label='Confirmed Cases', linewidth =2)
plot.plot(dateList,table['Cured'].tolist(),'m', label='Recovered Cases', linewidth =2)
plot.plot(dateList,table['Deaths'].tolist(),'r', label='Deceased Cases', linewidth =2)
          

plot.title('COVID-19 DATA IN JHARKHAND ')
plot.xlabel('Number Of Days ')
plot.ylabel('Number Of People')
plot.legend()
plot.xticks(rotation=90)
plot.grid(True)
plot.legend()
plot.savefig('pic.jpg')
plot.show()
```
